# Classical Functions (Dimesion = 100) [F1 - F13]

```python
classical_df = pd.read_csv(r'D:\ProfAjaz\Experiment#1\archived_result\new_result\results_f1_f13\benchmark_summary.csv')
```

```python
classical_df.columns.to_list()
```

```python
fitness_cols = ['Mean_Fitness', 'Std_Fitness']
percent_cols = ['Mean_Expl', 'Mean_Exploit', 'AUC_Expl', 'AUC_Exploit', 
                'Mean_Expl_Hussain', 'Mean_Exploit_Hussain', 'AUC_Expl_Hussain', 'AUC_Exploit_Hussain']
iter_cols = ['Avg_Conv_Iter', 'Avg_Xover_Point', 'Avg_Xover_Hussain']


for col in fitness_cols + percent_cols + iter_cols:
    classical_df[col] = pd.to_numeric(classical_df[col], errors='coerce')


summary_df = classical_df.groupby('Optimizer').agg({
    'Mean_Fitness': 'mean',
    'Std_Fitness': 'mean',
    'Mean_Expl': 'mean',
    'Mean_Exploit': 'mean',
    'AUC_Expl': 'mean',
    'AUC_Exploit': 'mean',
    'Mean_Expl_Hussain': 'mean',
    'Mean_Exploit_Hussain': 'mean',
    'Avg_Conv_Iter': 'mean',
    'Avg_Xover_Point': 'mean',
    'Avg_Xover_Hussain': 'mean'
}).reset_index()


display_df = summary_df.copy()

for col in fitness_cols:
    display_df[col] = display_df[col].apply(lambda x: f"{x:.2e}")


for col in percent_cols:
    if col in display_df.columns:
        display_df[col] = display_df[col].apply(lambda x: f"{x:.2f}")


for col in iter_cols:
    display_df[col] = display_df[col].round(0).astype(int)

print("--- SUMMARY AVERAGED (F1-F13 100D) ---")
print(display_df.to_string(index=False))
```

```python
pd.read_csv(r'D:\ProfAjaz\Experiment#1\archived_result\new_result\results_f1_f13\friedman_ranks.csv')
```

```python
classical_df[['Function', 'Optimizer', 'Mean_Expl', 'Mean_Exploit', 'AUC_Expl', 'AUC_Exploit', 'Mean_Expl_Hussain', 'Mean_Exploit_Hussain',
              'AUC_Expl_Hussain', 'AUC_Exploit_Hussain',	'Avg_Conv_Iter',	'Avg_Xover_Point',	'Avg_Xover_Hussain']]
```

```python
df_total = classical_df 


corr_val, p_val = stats.spearmanr(df_total['Mean_Expl'], df_total['Mean_Expl_Hussain'])

print(f"--- Spearman Correlation Results F1 - F13 (100D) ---")
print(f"Correlation Coefficient: {corr_val:.4f}")
print(f"P-value: {p_val:.4e}")
```
